**Create Dictionary and Corpus**

```python
# === 1) Load your Excel ===
# Change the path to your file:
PATH = "/content/drive/MyDrive/Colab Notebooks/alex/data/tokenized_newspaper_data_18_sep.xlsx"

# Only read the needed column to save RAM
tokenized_df = pd.read_excel(PATH)
```

```python
import pandas as pd
from gensim.corpora import Dictionary
from gensim.models import LdaModel, CoherenceModel
```

```python
import pandas as pd
from gensim.corpora import Dictionary
import ast

# Convert the string representation of lists back to actual lists, with error handling
texts = []
for item in tokenized_df['Main_text_tokenized']:
    if isinstance(item, str):
        try:
            texts.append(ast.literal_eval(item))
        except (SyntaxError, ValueError):
            texts.append([])  # Append empty list for malformed strings
    else:
        texts.append([]) # Append empty list for non-string values

```

```python
import json

# Save the texts list to a JSON file
with open('texts.json', 'w') as f:
    json.dump(texts, f)

print("Texts saved to 'texts.json'")
```

```python
dictionary = Dictionary(texts)

corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
import gensim
# Save the dictionary and corpus
dictionary.save('dictionary.gensim')
corpus_file = 'corpus.mm'
gensim.corpora.MmCorpus.serialize(corpus_file, corpus)

print(f"Dictionary saved to 'dictionary.gensim'")
print(f"Corpus saved to '{corpus_file}'")
```

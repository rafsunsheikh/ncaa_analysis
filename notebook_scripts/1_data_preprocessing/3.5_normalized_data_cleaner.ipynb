## Remove numbers from columns

### Subtask:
Define a function to remove numbers and apply it to the specified columns.

```python
import re

def remove_numbers(text):
    if isinstance(text, str):
        # Use regex to replace all digits with an empty string
        return re.sub(r'\d+', '', text).strip()
    return text

# Apply the function to the specified columns
new_normalized_data['title_normalized'] = new_normalized_data['title_normalized'].apply(remove_numbers)
new_normalized_data['main_text_normalized'] = new_normalized_data['main_text_normalized'].apply(remove_numbers)
```

# Task
Remove a list of words from the "title_normalized" and "main_text_normalized" columns.

## Define words to remove

### Subtask:
Create a list of words to be removed from the text columns.

```python
words_to_remove = ['v', 'pm', 'game', 'et', 'ppg', 'g', 'rpg', 'f', 'jr', 'apg', 'fg', 'sr', 'u']
```

## Remove words from columns

### Subtask:
Define a function to remove the words and apply it to the specified columns.

```python
import re

def remove_words(text, words):
    if isinstance(text, str):
        # Create a regex pattern to match whole words from the list
        pattern = r'\b(?:' + '|'.join(re.escape(word) for word in words) + r')\b'
        # Replace the matched words with an empty string
        return re.sub(pattern, '', text, flags=re.IGNORECASE).strip()
    return text

# Apply the function to the specified columns
new_normalized_data['title_normalized'] = new_normalized_data['title_normalized'].apply(lambda x: remove_words(x, words_to_remove))
new_normalized_data['main_text_normalized'] = new_normalized_data['main_text_normalized'].apply(lambda x: remove_words(x, words_to_remove))
```
